```python
import matplotlib.pyplot as plt
import numpy as np

def estimate_pi(N):
    adentro_circulo = 0
    x_adentro, y_adentro = [], []
    x_afuera, y_afuera = [], []
    
    for _ in range(N):
        x, y = np.random.uniform(-1, 1, 2)  
        if x**2 + y**2 <= 1:
            adentro_circulo += 1
            x_adentro.append(x)
            y_adentro.append(y)
        else:
            x_afuera.append(x)
            y_afuera.append(y)
    
    pi_estimate = (adentro_circulo / N) * 4
    return pi_estimate, x_adentro, y_adentro, x_afuera, y_afuera
```

```python
def plot_simulation(N):
    pi_estimate, x_adentro, y_adentro, x_afuera, y_afuera = estimate_pi(N)
    
    plt.figure(figsize=(6,6))
    plt.gca().add_patch(plt.Circle((0, 0), 1, color='blue', fill=False))
    
    plt.scatter(x_adentro, y_adentro, color='green', s=1, label='Dentro del círculo')
    plt.scatter(x_afuera, y_afuera, color='red', s=1, label='Fuera del círculo')
    
    plt.xlim(-1, 1)
    plt.ylim(-1, 1)
    plt.gca().set_aspect('equal')
    plt.legend()
    plt.title(f'N={N}\nπ ≈ {pi_estimate:.6f}')
    
    plt.show()
```

```python
for N in [10, 100, 1000, 10000, 100000]:
    plot_simulation(N)
```
